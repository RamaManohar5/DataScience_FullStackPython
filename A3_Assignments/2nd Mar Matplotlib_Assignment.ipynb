### Q2: What is a scatter plot? Use the following code to generate data for x and y. Using this generated data plot a scatter plot.
    
        Scatter plots are used to observe relationship between variables and uses dots to represent the relationship between them. 

```python
import matplotlib.pyplot as plt
import numpy as np
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))

plt.figure(figsize=(8,4))
plt.scatter(x, y, color = '#CE0D33', alpha = 0.9, marker = 'x')
plt.xlabel('x-axis data')
plt.ylabel('y-axis data')
plt.title('Scatter Plot')
plt.grid()
```

### Q3: Why is the subplot() function used? Draw four line plots using the subplot() function.
        
        subplot() method provides a way to plot multiple plots on a single figure. 
        

```python
import numpy as np
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 8))

# plot 1
x = np.array([0, 1, 2, 3, 4, 5])
y = np.array([0, 100, 200, 300, 400, 500])

plt.subplot(2, 2, 1)
plt.plot(x, y, color = '#CE0D33', alpha = 0.9)
plt.xlabel('x-axis data')
plt.ylabel('y-axis data')
plt.title("subplot 1", loc='left')
plt.grid()

# plot 2
x = np.array([0, 1, 2, 3, 4, 5])
y = np.array([50, 20, 40, 20, 60, 70])

plt.subplot(2, 2, 2)
plt.plot(x, y, color = '#CEB70D', alpha = 0.9)
plt.xlabel('x-axis data')
plt.ylabel('y-axis data')
plt.title("subplot 2", loc='left')
plt.grid()

# plot 3
x = np.array([0, 1, 2, 3, 4, 5])
y = np.array([10, 20, 30, 40, 50, 60])

plt.subplot(2, 2, 3)
plt.plot(x, y, color = '#A50DCE', alpha = 0.9)
plt.xlabel('x-axis data')
plt.ylabel('y-axis data')
plt.title("subplot 3", loc='left')
plt.grid()

# plot 4
x = np.array([0, 1, 2, 3, 4, 5])
y = np.array([200, 350, 250, 550, 450, 150])

plt.subplot(2, 2, 4)
plt.plot(x, y, color = '#0D99CE', alpha = 0.9)
plt.xlabel('x-axis data')
plt.ylabel('y-axis data')
plt.title("subplot 4", loc='left')
plt.grid()
```

### Q4: What is a bar plot? Why is it used? Using the following data plot a bar plot and a horizontal bar plot.

    A bar plot or bar chart is a graph that represents the category of data with rectangular bars with lengths and heights that is proportional to the values which they represent. The bar plots can be plotted horizontally or vertically. A bar chart describes the comparisons between the discrete categories. One of the axis of the plot represents the specific categories being compared, while the other axis represents the measured values corresponding to those categories.

```python
import numpy as np
import matplotlib.pyplot as plt

company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])

plt.figure(figsize=(5,3))
plt.bar(company, profit)
plt.xlabel('companies names')
plt.ylabel('profit')
plt.title("Bar Plot")
plt.grid()
```

```python
plt.figure(figsize=(6,3))
plt.barh(company, profit, color = '#CE0DB7' ) # horizontal bar
plt.ylabel('companies names')
plt.xlabel('profit')
plt.title("Bar Plot")
plt.grid()
```

### Q5: What is a box plot? Why is it used? Using the following data plot a box plot.
    
        A Box Plot is also known as Whisker plot is created to display the summary of the set of data values having properties like minimum, first quartile, median, third quartile and maximum. In the box plot, a box is created from the first quartile to the third quartile, a vertical line is also there which goes through the box at the median.

```python
import numpy as np
import matplotlib.pyplot as plt

box1 = np.random.normal(100, 10, 200)

plt.boxplot(box1)
plt.show()
```
