```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
```

```python
x = np.random.rand(50)
y = np.random.rand(50)
```

```python
x
```

```python
y
```

```python
plt.scatter(x, y)
```

```python

plt.figure(figsize=(6,2))
plt.scatter(x, y, c = '#9f61fd', alpha=0.9, marker='x')
plt.xlabel("this is my x_axis data")
plt.ylabel("this is my y_axis data")
plt.title("Scatter Plot")
plt.grid()
```

```python
x = np.linspace(1, 10, 100) # creates a dataset with equal line space in the given range

y = np.sin(x)
```

```python
x
```

```python
y
```

```python
plt.figure(figsize=(8, 2))
plt.plot(x, y,  "--r")
plt.xlabel("this is x_lable linespace data")
plt.ylabel("this is y_lable sin of x_lable data")
plt.title("Plot Graph - SinWave")
plt.grid()
```

```python
data = [1,2,3,3,5,3,5,3,8,8,6,5,4,9,3,5,6,7,8,9,9]
```

```python
plt.figure(figsize=(6,3))
plt.hist(data, color="#353903") # useful to see repetitive data, y - axis shows frequency
plt.xlabel("representing unique values")
plt.ylabel("representing frequency of values")
plt.title("Histogram")
plt.grid()
```
