```python
# merge & join operations

data1 = pd.DataFrame({"key1" : [2,3,4,5],
                      "key2" : [4,5,6,7],
                      "key3" : [8,9,10,11]
})
```

```python
data2 = pd.DataFrame({"key1" : [1,2,4,6,67],
                      "key4" : [56,5,6,7,8],
                      "key5" : [3,56,5,6,6]
})
```

```python
data1
```

```python
data2
```

```python
pd.merge(data1, data2) # it's an intersection based on common column value, bydefault innersection
```

```python
pd.merge(data1, data2, how = 'left') # left intersection
```

```python
pd.merge(data1, data2, how = 'outer')
```

```python
pd.merge(data1, data2, how = 'cross') # cross merge
```

```python
pd.merge(data1, data2, how = 'left', left_on='key2', right_on='key4')
```

```python
# join operation

data3 = pd.DataFrame({'key1' : [1,2,4,5,6],
                      'key2' : [4,5,6,7,8],
                      'key3' : [3,4,5,6,6]},
                     index = ['a', 'b', 'c', 'd', 'e']
                    )
```

```python

data4 = pd.DataFrame({'key11' : [1,2,4,5,6],
                      'key22' : [4,5,6,7,8],
                      'key33' : [3,4,5,6,6]},
                     index = ['a', 'b', 'h', 'i', 'j']
                    )
```

```python
data3
```

```python
data4
```

```python
data3.join(data4)
```

```python
data3.join(data4, how = 'inner')
```

```python
data3.join(data4, how = 'outer')
```

```python
data3.join(data4, how = 'cross')
```
