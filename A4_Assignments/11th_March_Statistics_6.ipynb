### Q1: What is the difference between a t-test and a z-test? Provide an example scenario where you would use each type of test.
    
    Both t-tests and z-tests are statistical tests used to determine if a sample mean is significantly different from a known population mean. However, they differ in the way they calculate the standard error and the test statistic used.

    A z-test assumes that the population standard deviation is known, while a t-test assumes that it is unknown and estimates it from the sample data. As a result, the t-test is more commonly used when the sample size is small or the population standard deviation is unknown.

    Here is an example scenario where each test would be appropriate:

    Scenario: A company wants to know if the average salary of their employees is significantly different from the average salary of all employees in their industry.

    Z-test: If the population standard deviation is known or assumed, the company can use a z-test to determine if the sample mean is significantly different from the population mean. For example, if the company knows that the population standard deviation is $10,000 and they sample 100 employees with a mean salary of $70,000, they can calculate a z-score and determine the p-value to test the hypothesis.

    T-test: If the population standard deviation is unknown, the company can use a t-test to estimate the standard error from the sample data. For example, if the company only has a sample of 20 employees with a mean salary of $70,000 and an estimated standard deviation of $12,000, they can use a t-test to determine if the sample mean is significantly different from the population mean.


### Q2: Differentiate between one-tailed and two-tailed tests.

    One-tailed and two-tailed tests are types of hypothesis tests used in statistics.

    A one-tailed test is a statistical test where the null hypothesis is rejected only if the sample mean is either significantly greater than or significantly less than the expected population mean in one specific direction. In other words, the hypothesis is tested only in one direction. A one-tailed test is often used when there is prior knowledge or a strong theoretical reason to expect the difference between the sample and the population to be in a particular direction.
    
    On the other hand, a two-tailed test is a statistical test where the null hypothesis is rejected if the sample mean is significantly different from the expected population mean in any direction, either greater than or less than. A two-tailed test is often used when there is no prior knowledge or theoretical reason to expect the difference between the sample and the population to be in a particular direction.

### Q3: Explain the concept of Type 1 and Type 2 errors in hypothesis testing. Provide an example scenario for each type of error.
    
    Type I error occurs when the null hypothesis is rejected, even though it is true. It is also known as a false positive. This error is represented by the alpha level of a test, which is the level of significance set by the researcher to reject the null hypothesis. A Type I error is more likely to occur when the alpha level is set at a high value.
    
    Example :
        In our clinical study, we compare the symptoms of patients who received the new drug intervention or a control treatment. Using a t test, we obtain a p value of .035. This p value is lower than our alpha of .05, so we consider our results statistically significant and reject the null hypothesis.
        However, the p value means that there is a 3.5% chance of our results occurring if the null hypothesis is true. Therefore, there is still a risk of making a Type I error.
    
    Type II error occurs when the null hypothesis is not rejected, even though it is false. It is also known as a false negative. This error is represented by the beta level of a test, which is the probability of failing to reject a false null hypothesis. A Type II error is more likely to occur when the sample size is small or the effect size is small.
    
    Example :
        When preparing our clinical study, we complete a power analysis and determine that with our sample size, we have an 80% chance of detecting an effect size of 20% or greater. An effect size of 20% means that the drug intervention reduces symptoms by 20% more than the control treatment.
        However, a Type II may occur if an effect that’s smaller than this size. A smaller effect size is unlikely to be detected in your study due to inadequate statistical power.

### Q4: Explain Bayes's theorem with an example.
        
    Bayes's theorem is a statistical rule used to calculate the probability of a hypothesis or event based on prior knowledge and new evidence. The theorem is named after the English mathematician and statistician Thomas Bayes.

    The formula for Bayes's theorem is as follows:

    P(A|B) = P(B|A) * P(A) / P(B)

    where:
    P(A|B) is the probability of event A occurring given that event B has occurred
    P(B|A) is the probability of event B occurring given that event A has occurred
    P(A) is the prior probability of event A
    P(B) is the prior probability of event B

    Here's an example of how to use Bayes's theorem:

    Suppose a medical test is used to diagnose a certain disease. The test has a 95% accuracy rate, meaning that if a person has the disease, there is a 95% chance that the test will come back positive. However, if a person does not have the disease, there is a 5% chance that the test will come back positive. The prevalence of the disease in the population is 1%.

    Using Bayes's theorem, we can calculate the probability that a person who tests positive actually has the disease:

    Let A be the event that a person has the disease
    Let B be the event that the test comes back positive

    P(A) = 0.01 (the prior probability of the disease in the population)
    P(B|A) = 0.95 (the probability of a positive test given the person has the disease)
    P(B|~A) = 0.05 (the probability of a positive test given the person does not have the disease)

    P(B) = P(B|A) * P(A) + P(B|~A) * P(~A)
    = 0.95 * 0.01 + 0.05 * 0.99
    = 0.058

    Now we can use Bayes's theorem to calculate the probability of having the disease given a positive test result:

    P(A|B) = P(B|A) * P(A) / P(B)
    = 0.95 * 0.01 / 0.058
    = 0.163

    Therefore, a person who tests positive for the disease has a 16.3% chance of actually having the disease, despite the high accuracy rate of the test. This example illustrates the importance of considering the prior probability of an event when interpreting the results of a test or study.


### Q5: What is a confidence interval? How to calculate the confidence interval, explain with an example.
    
    A confidence interval is a range of values that is likely to contain the true value of a population parameter with a certain degree of confidence, based on a sample of data. It is an estimate of the precision and uncertainty of the sample estimate.
    
    we can use the following formula to calculate the confidence interval:

    CI = X ± z * (s / sqrt(n))

    where:
    CI is the confidence interval
    X is the sample mean
    z is the critical value for the desired level of confidence
    s is the sample standard deviation
    n is the sample size
    
    Ex:
    Suppose we have a random sample of 50 students and their scores on a math test. The sample mean is 75 and the sample standard deviation is 10. We want to estimate the population mean with 95% confidence.
    CI = 75 ± 1.96 * (10 / sqrt(50))
        = 75 ± 2.78
    
    Therefore, the 95% confidence interval for the population mean is (72.22, 77.78). This means that if we were to repeat this study multiple times, we would expect the true population mean to fall within this interval 95% of the time.

### Q6. Use Bayes' Theorem to calculate the probability of an event occurring given prior knowledge of the event's probability and new evidence. Provide a sample problem and solution.

    Suppose a certain medical test is used to diagnose a rare disease, which affects only 0.1% of the population. The test has a sensitivity of 95%, meaning that it correctly identifies 95% of people who have the disease, and a specificity of 90%, meaning that it correctly identifies 90% of people who do not have the disease. If a person tests positive for the disease, what is the probability that they actually have it?

    Let's use Bayes' Theorem to calculate this probability:

    Let A be the event that the person has the disease, and B be the event that the test is positive.

    P(A) = 0.001, the prior probability of the person having the disease
    P(~A) = 0.999, the prior probability of the person not having the disease
    P(B|A) = 0.95, the probability of a positive test given that the person has the disease
    P(B|~A) = 0.10, the probability of a positive test given that the person does not have the disease

    We can use the formula for Bayes' Theorem to calculate P(A|B), the probability of the person having the disease given that the test is positive:

    P(A|B) = P(B|A) * P(A) / P(B)
    = P(B|A) * P(A) / [P(B|A) * P(A) + P(B|~A) * P(~A)]
    = 0.95 * 0.001 / [0.95 * 0.001 + 0.10 * 0.999]
    = 0.008

    Therefore, the probability of the person actually having the disease given a positive test result is only 0.8%. This illustrates how even highly accurate tests can lead to a relatively low probability of a positive result indicating the presence of a rare condition.

It is calculated by multiplying the critical value (based on the desired level of confidence) by the standard error of the sample.

    The sample size affects the margin of error in two ways. First, a larger sample size decreases the standard error of the sample, which reduces the margin of error. Second, a larger sample size allows for a more accurate representation of the population, which also reduces the margin of error.

    For example, let's say we want to estimate the proportion of voters in a particular state who support a specific candidate. ### Q9. Calculate the z-score for a data point with a value of 75, a population mean of 70, and a population standard deviation of 5. Interpret the results.

    z = (x - μ) / σ

    Where:
        x = data point value
        μ = population mean
        σ = population standard deviation

    Using the given values, we can calculate the z-score as follows:

        z = (75 - 70) / 5
        z = 1

    So the z-score for the data point with a value of 75 is 1. This means that the data point is 1 standard deviation above the population mean.

    Interpreting the z-score, we can say that the data point of 75 is above average and relatively far from the population mean. Specifically, the data point is one standard deviation above the mean, which is a relatively common occurrence, as about 68% of all data points fall within one standard deviation of the mean in a normal distribution.

### Q10. In a study of the effectiveness of a new weight loss drug, a sample of 50 participants lost an average of 6 pounds with a standard deviation of 2.5 pounds. Conduct a hypothesis test to determine if the drug is significantly effective at a 95% confidence level using a t-test.
    
    To conduct a hypothesis test, we need to set up our null and alternative hypotheses.

    Null Hypothesis (H0): The weight loss drug is not significantly effective, and the average weight loss in the population is equal to or less than 0 pounds.

    Alternative Hypothesis (Ha): The weight loss drug is significantly effective, and the average weight loss in the population is greater than 0 pounds.

    We will use a one-tailed t-test with a 95% confidence level to determine if we can reject the null hypothesis.

    The formula for the t-test is:

    t = (x - μ) / (s / sqrt(n))

    Where:
        x = sample mean
        μ = hypothesized population mean
        s = sample standard deviation
        n = sample size

    Using the given values, we can calculate the t-value as follows:

        t = (6 - 0) / (2.5 / sqrt(50))
        t = 15.56

    The critical value for a one-tailed t-test with 49 degrees of freedom (50 participants - 1) at a 95% confidence level is 1.676. Since our calculated t-value (15.56) is greater than the critical value (1.676), we can reject the null hypothesis and conclude that the weight loss drug is significantly effective at a 95% confidence level.

    Therefore, we can conclude that the weight loss drug is significantly effective in causing weight loss in the population.

### Q11. In a survey of 500 people, 65% reported being satisfied with their current job. Calculate the 95% confidence interval for the true proportion of people who are satisfied with their job.
    
    To calculate the 95% confidence interval for the true proportion of people who are satisfied with their job, we can use the following formula:

    CI = p ± z* (sqrt(p * q / n))

    Where:
        CI = confidence interval
        p = sample proportion
        q = 1 - p (the complement of the sample proportion)
        n = sample size
        z* = z-value for the desired confidence level (95% in this case)

    Given the information in the question, we can plug in the values as follows:

        p = 0.65
        q = 1 - p = 0.35
        n = 500
        z* = 1.96 (for a 95% confidence level)

    Now, we can calculate the confidence interval:

        CI = 0.65 ± 1.96 * (sqrt(0.65 * 0.35 / 500))
        CI = 0.65 ± 0.0455
        CI = [0.6045, 0.6955]

    Therefore, we can be 95% confident that the true proportion of people who are satisfied with their job falls within the interval of [0.6045, 0.6955].
    

### Q12. A researcher is testing the effectiveness of two different teaching methods on student performance. Sample A has a mean score of 85 with a standard deviation of 6, while sample B has a mean score of 82 with a standard deviation of 5. Conduct a hypothesis test to determine if the two teaching methods have a significant difference in student performance using a t-test with a significance level of 0.01.
    
        To conduct a hypothesis test to determine if there is a significant difference in student performance between the two teaching methods, we need to set up our null and alternative hypotheses:

    Null Hypothesis (H0): The mean scores of the two teaching methods are not significantly different.
    Alternative Hypothesis (Ha): The mean scores of the two teaching methods are significantly different.

    We can use a two-tailed t-test with a significance level of 0.01 to test our hypotheses.

    The formula for the t-test for two independent samples is:

    t = (x1 - x2) / sqrt((s1^2/n1) + (s2^2/n2))

    Where:
        x1 and x2 are the sample means
        s1 and s2 are the sample standard deviations
        n1 and n2 are the sample sizes
    
    Using the given values, we can calculate the t-value as follows:

        t = (85 - 82) / sqrt((6^2/1) + (5^2/1))
        t = 3 / sqrt(61)
        t = 0.387

    The degrees of freedom for this test is (n1 - 1) + (n2 - 1) = 1 + 1 = 2.

    The critical t-value for a two-tailed test with 2 degrees of freedom and a significance level of 0.01 is ±6.965. Since our calculated t-value (0.387) is less than the critical value, we fail to reject the null hypothesis and conclude that there is not a significant difference in student performance between the two teaching methods.

    Therefore, we can conclude that there is insufficient evidence to suggest that one teaching method is more effective than the other based on the sample data.

### Q14. In a study of the effects of caffeine on reaction time, a sample of 30 participants had an average reaction time of 0.25 seconds with a standard deviation of 0.05 seconds. Conduct a hypothesis test to determine if the caffeine has a significant effect on reaction time at a 90% confidence level using a t-test.
    
    To conduct a hypothesis test to determine if caffeine has a significant effect on reaction time, we need to set up our null and alternative hypotheses:

    Null Hypothesis (H0): Caffeine has no significant effect on reaction time.
    Alternative Hypothesis (Ha): Caffeine has a significant effect on reaction time.

    We can use a one-sample t-test with a significance level of 0.1 (90% confidence level) to test our hypotheses.

    The formula for the t-test for a single sample is:

        t = (x - μ) / (s / sqrt(n))

    Where:
        x is the sample mean
        μ is the hypothesized population mean under the null hypothesis
        s is the sample standard deviation
        n is the sample size

    Using the given values, we can calculate the t-value as follows:

        t = (0.25 - μ) / (0.05 / sqrt(30))

    To determine the critical value of t, we need to use a t-distribution table with degrees of freedom (df) = n - 1 = 30 - 1 = 29 and a significance level of 0.1. The critical t-value is ±1.699.
        
        If the calculated t-value is greater than the critical value, we reject the null hypothesis and conclude that caffeine has a significant effect on reaction time. If the calculated t-value is less than or equal to the critical value, we fail to reject the null hypothesis.

    To solve for μ, we rearrange the formula to solve for μ:

        μ = x - (t * s / sqrt(n))

    Plugging in the given values and the critical t-value, we get:

        μ = 0.25 - (1.699 * 0.05 / sqrt(30))
        μ = 0.231

    Therefore, the null hypothesis is:

        H0: μ = 0.231

    Now, we can calculate the t-value:

        t = (0.25 - 0.231) / (0.05 / sqrt(30))
        t = 1.8

    Since the calculated t-value (1.8) is greater than the critical value (1.699), we can reject the null hypothesis and conclude that caffeine has a significant effect on reaction time at a 90% confidence level.

    Therefore, we can conclude that based on the sample data, there is sufficient evidence to suggest that caffeine has a significant effect on reaction time.
